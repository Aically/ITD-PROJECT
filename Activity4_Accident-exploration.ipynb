<h1>Insight no. 7</h1>

<h1>Is the accident happen between year 2019 and 2020 increase </h1>

<h1>Analyzing the accident severity data from 2019 and 2020 reveals notable trends:

2019 Data: The dataset includes 1,752 fatalities, 25,945 serious injuries, and 125,461 slight injuries, totaling 153,158 casualties.

2020 Data: There were 1,460 fatalities, 20,102 serious injuries, and 92,055 slight injuries, amounting to 115,584 casualties.</h1>

```python
severity_counts_2019 = df_year19['Accident_Severity'].value_counts()
severity_counts_2020 = df_year20['Accident_Severity'].value_counts()
```

```python
all_severities = severity_counts_2019.index.union(severity_counts_2020.index)
severity_counts_2019 = severity_counts_2019.reindex(all_severities, fill_value=0)
severity_counts_2020 = severity_counts_2020.reindex(all_severities, fill_value=0)
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.35  
x = range(len(all_severities))
ax.bar(x, severity_counts_2019, width, label='2019', color='skyblue')
ax.bar([p + width for p in x], severity_counts_2020, width, label='2020', color='salmon')
ax.set_xlabel('Accident Severity')
ax.set_ylabel('Number of Accidents')
ax.set_title('Comparison of Accident Severity: 2019 vs 2020')
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(all_severities, rotation=45)
ax.legend()
plt.tight_layout()
plt.show()
```
